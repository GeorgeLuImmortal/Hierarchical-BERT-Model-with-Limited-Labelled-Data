```python
from scipy import stats
import scikit_posthocs as sp
import numpy as np
from scipy.stats import wilcoxon
import pandas as pd
```

```python
headers = ['FastText','RoBERTa','Fine-tuned RoBERTa','HAN','HBM']

def Nemenyi(headers:list,data:list):
    data = np.array(data)
    df_nemenyi = sp.posthoc_nemenyi_friedman(data.T)
    df_nemenyi = df_nemenyi.rename(columns={0: headers[0],1: headers[1],2: headers[2],3: headers[3],4: headers[4]},index={0: headers[0],1: headers[1],2: headers[2],3: headers[3],4: headers[4]})
    
    return df_nemenyi

def Wilcoxon(headers:list,data:list):
    
    p_matrix= [] 
    for idx,row in enumerate(data):
        ps = []
        for i in range(idx+1,len(data)):
            pvalue = wilcoxon(row,data[i])[1]
            print('P value of %s and %s is:'%(headers[idx],headers[i]),pvalue)
            ps.append(pvalue)
        p_matrix.append(ps)
        
    standard_pmatrix = []
    for ps in p_matrix:
        if len(ps)<=5:
            ps =[0.0]*(5-len(ps))+ ps
            standard_pmatrix.append(ps)
    
    standard_pmatrix = np.array(standard_pmatrix) 
    standard_pmatrix = standard_pmatrix + standard_pmatrix.T - np.diag(np.diag(standard_pmatrix))
    np.fill_diagonal(standard_pmatrix, -1)
            
    df_wilcoxon = pd.DataFrame(standard_pmatrix)
    df_wilcoxon = df_wilcoxon.rename(columns={0: headers[0],1: headers[1],2: headers[2],3: headers[3],4: headers[4]},index={0: headers[0],1: headers[1],2: headers[2],3: headers[3],4: headers[4]})
                                     
    return df_wilcoxon
```
